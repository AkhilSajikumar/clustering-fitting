```python
# Assuming Card_df is your DataFrame containing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(Card_df)

# Initialize an empty list to store within-cluster sum of squares
wcss = []

# Fit KMeans for different values of K
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plotting the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# Assuming Card_df is your DataFrame containing the data
# Replace 'Card_df' with the actual DataFrame name if it's different
features = Card_df[['Avg_Credit_Limit', 'Total_Credit_Cards', 'Total_visits_bank', 'Total_visits_online', 'Total_calls_made']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Initialize KMeans with the desired number of clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

# Fit KMeans to the scaled features
kmeans.fit(scaled_features)

# Add cluster labels to the DataFrame
Card_df['Cluster'] = kmeans.labels_

# Display the cluster centers
print("Cluster Centers:")
print(pd.DataFrame(kmeans.cluster_centers_, columns=features.columns))

# Display the count of customers in each cluster
print("\nCount of customers in each cluster:")
print(Card_df['Cluster'].value_counts())
plt.show()
```

```python
# Assuming Card_df is your DataFrame containing the data
# Replace 'Card_df' with the actual DataFrame name if it's different
features = Card_df[['Avg_Credit_Limit', 'Total_Credit_Cards', 'Total_visits_bank', 'Total_visits_online', 'Total_calls_made']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(scaled_features)

# Add cluster labels and centroids to the DataFrame
Card_df['Cluster'] = kmeans.labels_
Card_df['Centroid'] = kmeans.predict(scaled_features)

# Calculate centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

# Plot the clusters
plt.figure(figsize=(10, 6))
for cluster in range(5):  # Assuming 5 clusters
    plt.scatter(Card_df.loc[Card_df['Cluster'] == cluster, 'Avg_Credit_Limit'], 
                Card_df.loc[Card_df['Cluster'] == cluster, 'Total_calls_made'], 
                label=f'Cluster {cluster}', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 4], color='black', marker='x', label='Centroids')
plt.xlabel('Average Credit Limit')
plt.ylabel('Total Calls Made')
plt.title('Cluster Plot with Centroids')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the function for fitting the curve
def func(x, a, b, c):
    return a * x**2 + b * x + c

# Extract data from the DataFrame
avg_credit_limit = Card_df['Avg_Credit_Limit']
total_calls_made = Card_df['Total_calls_made']

# Fit the curve
popt, pcov = curve_fit(func, avg_credit_limit, total_calls_made)

# Generate data points for the fitted curve
x_fit = np.linspace(min(avg_credit_limit), max(avg_credit_limit), 100)
y_fit = func(x_fit, *popt)

# Plot the original data and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(avg_credit_limit, total_calls_made, label='Original Data')
plt.plot(x_fit, y_fit, color='red', label='Fitted Curve')
plt.xlabel('Average Credit Limit')
plt.ylabel('Total Calls Made')
plt.title('Curve Fitting: Average Credit Limit vs Total Calls Made')
plt.legend()
plt.grid(True)
plt.show()
```
